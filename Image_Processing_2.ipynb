## Question 1

Read an image in the given link as ‘aaa.png’. Use the dilation operation on this image with two different structuring elements (square and circle). This size of structuring element is of your choice. Save your results as result-1.png andresult-2.png. Compare and contrast your results. Explain your observations.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('aaa.png', cv2.IMREAD_GRAYSCALE)
image_np = np.array(image)

square = cv2.resize(cv2.imread('square_structure.png',cv2.IMREAD_GRAYSCALE),(5,5))
circle = cv2.resize(cv2.imread('circle_structure.png',cv2.IMREAD_GRAYSCALE),(5,5))

# Perform dilation
result1 = cv2.resize(cv2.dilate(image, square, iterations=1),None ,fx=2,fy=2)
result2 = cv2.resize(cv2.dilate(image, circle, iterations=1),None, fx=2,fy=2)

plt.figure(figsize=(15, 10))


image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
result1 = cv2.cvtColor(result1, cv2.COLOR_GRAY2RGB)
result2 = cv2.cvtColor(result2, cv2.COLOR_GRAY2RGB)
# Original Image
plt.subplot(3, 1, 1)
plt.imshow(image)
plt.title('Original')
plt.axis('off')

# Result 1
plt.subplot(3, 1, 2)
plt.imshow(result1)
plt.title('Result 1 - square structure')
plt.axis('off')

# Result 2
plt.subplot(3, 1, 3)
plt.imshow(result2)
plt.title('Result 2 - Circle structure')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Both the structuring elements tend to close the holes in the image because we dilate the image using.

However square structuring element tends to make the hole more boxier, while the circle structuring element makes it more circle. Not only the holes, but the square structuring makes all the shapes get a sharp edge, while circle makes it more round.

## Question 2

Read an image. Add salt/pepper noise to it. Use a suitable structuring element to dilate this image, hence removing the noise from the image. Save your image as denoised.png.

```python
def add_noise(image, salt_prob, pepper_prob):
  noisy_image = np.copy(image)
  salt_mask = np.random.rand(*image.shape) < salt_prob
  pepper_mask = np.random.rand(*image.shape) < pepper_prob

  noisy_image[salt_mask] = 255
  noisy_image[pepper_mask] = 0
  return noisy_image
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rick = cv2.imread('Taj-Mahal.jpg',cv2.IMREAD_GRAYSCALE)
circle = cv2.imread('circle_structure.png',cv2.IMREAD_GRAYSCALE)
noisy_image = add_noise(rick,0.1,0.1)
noisy_image_cv = np.array(rick, dtype=np.uint8)
dilated_image = cv2.dilate(noisy_image_cv, cv2.resize(circle,(2,2)), iterations=1)

plt.figure(figsize=(15, 10))

plt.subplot(2,1,1)
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(2,1,2)
plt.imshow(dilated_image, cmap='gray')
plt.title('Dilated Image - structured using circle')
plt.axis('off')

plt.show()
```

## Question 6

Use the opening technique on the image available at the following link with the appropriate structuring element.to separate the circles from the lines. Save your image as circle.png and line.png. The circle_png image should contain only circles, and the line.png image should contain only lines. Show the structuring element used. Experiment with many structuring elements, describe your results and observations.

```python
def opening(image, structure):
  eroded = cv2.erode(image, structure)
  return cv2.dilate(eroded, structure)
```

## Circles

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

stick_and_circle = cv2.imread('8aCNN.jpg', cv2.IMREAD_GRAYSCALE)
circle_struct = cv2.imread('circle_structure.png', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 10))

plt.subplot(1,2,1)
plt.imshow(stick_and_circle, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(circle_struct, cmap='gray')
plt.title('Structuring element')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, size in enumerate(range(4,25,4)):
  circle_struct_resized = cv2.resize(circle_struct,(size,size))
  only_circles = opening(stick_and_circle, circle_struct_resized)
  plt.subplot(2, 3, i+1)
  plt.imshow(only_circles, cmap='gray')
  plt.title(f'Size {size}')
  plt.axis('off')

plt.show()
```

Thus a circle of dimension 16X16 is most suited to seperate the circles.

## Lines

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

size = 20
square_struct_1 = np.ones((1, size))
square_struct_2 = np.ones((size,1))
square_struct_3 = np.identity(size).astype(np.uint8)
square_struct_4 = np.identity(size).astype(np.uint8)[::-1, :]
only_sticks_1 = opening(stick_and_circle, square_struct_1)
only_sticks_2 = opening(stick_and_circle, square_struct_2)
only_sticks_3 = opening(stick_and_circle, square_struct_3)
only_sticks_4 = opening(stick_and_circle, square_struct_4)
only_sticks = only_sticks_1 + only_sticks_2 + only_sticks_3 + only_sticks_4

plt.figure(figsize=(10, 10))

plt.subplot(2,3,1)
plt.imshow(stick_and_circle, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(square_struct_1, cmap='gray')
plt.title('Structuring element 1')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(square_struct_2, cmap='gray')
plt.title('Structuring element 2')
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(square_struct_3, cmap='gray')
plt.title('Structuring element 3')
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(square_struct_4, cmap='gray')
plt.title('Structuring element 4')
plt.axis('off')

plt.subplot(2,3,6)
plt.imshow(only_sticks, cmap='gray')
plt.title('Only Sticks')
plt.axis('off')

plt.show()
```

4 Structuring elements are needed for the line, that is 1 horizontal line, 1 vertical line and 2 diagonals in opposite direction

## Question 7

Demonstrate Closing operation with any image of your choice and record your observations.

```python
def closing(image, struct):
  dilated = cv2.dilate(image,struct)
  return cv2.erode(dilated, struct)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

input = cv2.imread('8aCNN.jpg', cv2.IMREAD_GRAYSCALE)
circle_struct = cv2.resize(cv2.imread('circle_structure.png', cv2.IMREAD_GRAYSCALE),(16,16))

output = closing(input, circle_struct)
plt.figure(figsize=(10, 10))

plt.subplot(1,2,1)
plt.imshow(input, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(output, cmap='gray')
plt.title('Closed Image')
plt.axis('off')

plt.show()
```

Closing has caused new connections to form between different shapes
